```python
        #Mars Mission Planner

# Speeds in km/h for each spacecraft
speed_pathfinder = 40_000     # Pathfinder
speed_perseverance = 75_000   # Perseverance
speed_starship = 120_000      # Starship

# Mars average distance from Earth in kilometers
mars_distance_km = 225_000_000  

# Fuel usage and price data
fuel_usage_per_km = 0.3    # liters burned per km
fuel_price_per_liter = 1.80  # dollars per liter

# Function definitions
def calculate_travel_time(distance_km, speed_kmh):
    """
    Calculate travel time to Mars at a given speed.
    Formula: time (hours) = distance ÷ speed
    Rounds the result to the nearest hour.
    """
    hours = distance_km / speed_kmh
    return round(hours)

def calculate_mission_cost(distance_km, fuel_usage, fuel_price):
    """
    Calculate total fuel cost for the mission.
    Formula: total cost = distance × fuel usage × price per liter
    """
    total_fuel_needed = distance_km * fuel_usage
    total_cost = total_fuel_needed * fuel_price
    return total_cost
    
# Calculations for each mission

# Pathfinder
pathfinder_time = calculate_travel_time(mars_distance_km, speed_pathfinder)
pathfinder_cost = calculate_mission_cost(mars_distance_km, fuel_usage_per_km, fuel_price_per_liter)

# Perseverance
perseverance_time = calculate_travel_time(mars_distance_km, speed_perseverance)
perseverance_cost = calculate_mission_cost(mars_distance_km, fuel_usage_per_km, fuel_price_per_liter)

# Starship
starship_time = calculate_travel_time(mars_distance_km, speed_starship)
starship_cost = calculate_mission_cost(mars_distance_km, fuel_usage_per_km, fuel_price_per_liter)

# Mission Report Output
print("=====  Mars Mission Report =====\n")

print(f"--- Mission: Pathfinder ---")
print(f"Travel time (hours): {pathfinder_time}")
print(f"Total fuel cost: ${pathfinder_cost:,.2f}\n")

print(f"--- Mission: Perseverance ---")
print(f"Travel time (hours): {perseverance_time}")
print(f"Total fuel cost: ${perseverance_cost:,.2f}\n")

print(f"--- Mission: Starship ---")
print(f"Travel time (hours): {starship_time}")
print(f"Total fuel cost: ${starship_cost:,.2f}")
```
